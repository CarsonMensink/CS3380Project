```python
from faker import Faker
import random
import sqlite3

fake = Faker()
conn = sqlite3.connect("university_database.db")
cursor = conn.cursor()

cursor.execute("PRAGMA foreign_keys = ON;")

print("Connection successful")
```

```python
def populate_departments():
  departments = [
    "Computer Science",
    "Biology",
    "Mathematics",
    "Physics",
    "Chemistry"
  ]

  for department in departments:
    cursor.execute("""
    INSERT OR IGNORE INTO Department (department_name) VALUES (?);
    """, (department,))

  conn.commit()
  print("Departments populated.")
```

```python
def populate_courses():
  courses = [
    ("CMP_SC 3380", "Database Systems", 3, "Computer Science"),
    ("CMP_SC 2050", "Discrete Structures", 3, "Computer Science"),
    ("BIO_SC 1500", "General Biology", 4, "Biology"),
    ("MATH 2300", "Calculus I", 5, "Mathematics"),
    ("PHYSCS 2750", "University Physics", 5, "Physics"),
    ("CHEM 1320", "General Chemistry", 5, "Chemistry")
  ]

  for course_number, course_title, credits, department_name in courses:
    cursor.execute("""
    SELECT department_id
    FROM Department
    WHERE department_name = ?;
    """, (department_name,))

    department_id = cursor.fetchone()[0]

    cursor.execute("""
    INSERT OR IGNORE INTO Course (
      course_number,
      course_title,
      credits,
      department_id
    )
    VALUES (?, ?, ?, ?);
    """, (course_number, course_title, credits, department_id))

  conn.commit()
  print("Courses populated.")
```

```python
def populate_students(number_of_students=100):
    for _ in range(number_of_students):
        cursor.execute("""
        INSERT INTO Student (
            first_name,
            last_name,
            email
        )
        VALUES (?, ?, ?);
        """, (
            fake.first_name(),
            fake.last_name(),
            fake.unique.email()
        ))

    conn.commit()
    print("Students populated.")
```

```python
populate_departments()
populate_courses()
populate_students()
```
